### 6. Phase osztály

A `Phase` osztály főként az FT és a WFT kiértékelési módszerek miatt jött létre, itt ugyanis nem feltétlen szükséges kikötni, hogy a fázis polinommal közelíthető. A `GD_mode` argumentumtól függően használható fázis és GD görbe elemzésére is. Ilyen fázis objektumot kapunk meg, amikor a `build_phase` vagy a `build_GD` metódusokat használjuk. Nézzük meg egy egyszerű példán keresztül:

```python
import numpy as np
import matplotlib.pyplot as plt
from pysprint.core.phase import Phase
```

Ehhez itt egy tetszőleges polinomot használok:

```python
x = np.linspace(1, 30, 2000)
y = 2 * x ** 3 - 3 * x ** 2 + 45 * x + 3

fazis = Phase(x, y)
fazis.plot()
```

A `fit` függvénnyel ráillesztek egy harmadfokú görbét:

```python
fazis.fit(reference_point=0, order=3);
```

Itt látható, hogy a megadott együtthatókat adta vissza a megfelelő faktoriálissal szorozva.
Ha ugyan ezt `GD_mode=True` mellett végzem el:

```python
GD = Phase(x, y, GD_mode=True)
GD.fit(reference_point=0, order=4);
```

Itt megjegyezném, hogy az `order` kulcsszó minden esetben a **keresett diszperzió rendjét** és nem feltétlenül az illesztés rendjét jelenti. Ennek köszönhető, hogy a harmadfokú polinomra `GD_mode=True` mellett az `order=4` volt a megfelelő, hiszen ekkor kerül harmadfokú görbe illeszésre.

```python
GD.plot()
```

A hibák az `errorplot` függvénnyel nézhetők meg, a hibák mint `np.ndarray` az `errors` kulcsszóval érhetők el:

```python
GD.errorplot(percent=True)
```

```python
GD.errors
```

A `slice`, vagyis a kivágás ugyan úgy működik, mint minden kiértékelés esetén:

```python
fazis3 = Phase(x, y)

fig, ax = plt.subplots(figsize=(8, 8))

fazis3.plot(ax=ax, color="black", label="kivágás előtt")

fazis3.slice(5, 25)

fazis3.plot(ax=ax, color="red", label="kivágás után")
plt.grid()
plt.legend()
```

Az objektumban tárolt adatok a `data` függvénnyel érhetők el:

```python
x, y = fazis3.data
print(x, type(x))
print(y, type(y))
```

Természetesen a hibák nem léteznek addig, amíg nem illesztettünk görbét:

```python
fazis3.errors
```
