# Generator


The Generator class - as its name suggests - allows to simulate interferograms. Every unit is fs.

```python
%matplotlib inline

from pysprint.api import Generator

#setting up parameters
g = Generator(2.2, 2.8, 2.5, delay = 10000, GDD = 2000, FOD = -100000, normalize = True)
#chosing the domain
g.generate_freq()
#plotting the spectrogram and phase
g.phase_graph()
```

You can unpack the generated values via

```python
x, y, ref, sam = g.unpack()
```

Note: if normalize is True unpack returns 4-tuple, else it returns a 2-tuple.

### Accessing the plot

Generator objects have *show* method, which plots the current generated spectrogram. You can access the matplotlib figure by calling the *plotwidget* method.

```python
g.plotwidget.xlabel('Frequency [PHz]')
g.plotwidget.ylabel('Normalized intensity')
g.show() # make sure you call show at the end, otherwise things can get messed up.
```

### Saving:
Despite you can use *np.savetxt*, Generator has a built-in save method.

```python
g.save('myspectrum.txt') #Will save to the current working directory
# g.save('myspectrum.txt', path = 'C://...') --- > the save location can be changed through path argument. 
```

For more detailed descriptions you can call the *help()* on the desired class or method.
