# Generator


The Generator class allows to simulate interferograms.

The units are usually in `nm` and `fs`, and it's according higher powers. 

```python
%matplotlib inline

from pysprint import Generator

g = Generator(start=2.2, stop=2.8, center=2.5, delay=100, GDD=2000, FOD=-100000, normalize=True)

# generating in frequency domain
g.generate_freq()

#plotting the spectrogram and phase together
g.phase_graph()
```

You can access the `pysprint.Generator`'s data easily via

```python
x, y, ref, sam = g.data
```

Note: if normalize is `True` data will return 4-element tuple, else it returns a 2-element tuple.

### Accessing the plot

Generator objects have `show` method, which will plot the generated data and call `matplotlib.pyplot.show`.
To access the usual matplotlib functionality, you can just do the following:

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 6))

plt.xlabel('Frequency [PHz]')
plt.ylabel('Normalized intensity')

# calling g.show() instead of plt.show()
g.show() 
```

### Saving:
Despite you can use *np.savetxt*, Generator has a built-in save method.

```python
g.save('myspectrum.txt')
```
